```python
import os
import time
```

```python
def clearScreen():
    time.sleep(2)
    os.system('cls')
def askNumbers():
  try:
    number1 = float(input())
    number2 = float(input())

  except ValueError:
    print("Invalid Input")
    return None, None



  return number1, number2

def add():
  number1,number2 = askNumbers()
  print(number1,'+',number2,'=',number1+number2)

def sub():
  number1,number2 = askNumbers()
  print(number1,'-',number2,'=',number1-number2)

def divide():
  number1,number2 = askNumbers()
  if number2 == 0:
      print("Mathematical Error")
  else:
    print(number1,'/',number2,'=',round(number1/number2,3))

def mul():
  number1,number2 = askNumbers()
  print(number1,'X',number2,'=',number1*number2)

def remainder():
  number1,number2 = askNumbers()
  print(number1,'%',number2,'=',number1%number2)



```

```python
loop = True
while loop:
  operations = ['Add', 'Subtract', 'Multiply', 'Divide','Remainder', 'Exit']
  i = 1
  for operation in operations:
    print(i,')',operation)
    i+=1

  option = input("Enter Operation Number:\a")
  match option:
      case '1':
        add()
      case '2':
        sub()
      case '3':
        mul()
      case '4':
        divide()
      case '5':
        remainder()
      case '6':
        loop = False
        exit()
      case _ :
        print("Invalid Input")
        clearScreen()

```
