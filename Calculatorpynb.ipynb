```python
import os
import time
import operator
```

```python
def clearScreen():
    time.sleep(2)
    os.system('cls')

def askNumbers():
  try:
    number1 = float(input("Enter Number: "))
    number2 = float(input("Enter Number: "))

  except ValueError:
    print("Invalid Input")
    return None, None
  return number1, number2

  if number2 == 0:
      print("Mathematical Error")


```

```python
loop = True
while loop:
  operations = ['Add', 'Subtract', 'Multiply', 'Divide','Remainder', 'Exit']
  i = 1
  for operation in operations:
    print(i,')',operation)
    i+=1

  option = input("Enter Operation Number:\a")
  match option:
      case '1':
        number1,number2 = askNumbers()
        print(number1,'+',number2,'=',operator.add(number1,number2))
      case '2':
        number1,number2 = askNumbers()
        print(number1,'-',number2,'=',operator.sub(number1,number2))
      case '3':
        number1,number2 = askNumbers()
        print(number1,'X',number2,'=',operator.mul(number1,number2))
      case '4':
        number1,number2 = askNumbers()
        if number2 == 0:
          print("Mathematical Error")
        else:
          print(number1,'/',number2,'=',operator.truediv(number1,number2))
      case '5':
        number1,number2 = askNumbers()
        print(number1,'%',number2,'=',operator.mod(number1,number2))
      case '6':
        loop = False
        exit()
      case _ :
        print("Invalid Input")
        clearScreen()

```
